```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS asteroids (

    id BIGINT,
    name VARCHAR(200),
    absolute_magnitude_h FLOAT,
    estimated_diameter_min_km FLOAT,
    estimated_diameter_max_km FLOAT,
    is_potentially_hazardous_asteroid BOOLEAN,

    PRIMARY KEY(id)

);
""")

print("Asteroids Table Created")
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS close_approach (

    approach_id INT AUTO_INCREMENT PRIMARY KEY,

    neo_reference_id BIGINT,

    close_approach_date DATE,

    relative_velocity_kmph FLOAT,

    astronomical FLOAT,

    miss_distance_km FLOAT,

    miss_distance_lunar FLOAT,

    orbiting_body VARCHAR(30),

    FOREIGN KEY (neo_reference_id)
    REFERENCES asteroids(id)

);
""")

print("Close Approach Table Created")
```

Query 1
Count how many times each asteroid has approached Earth

```python
query = """
SELECT
    a.name,
    COUNT(*) AS total_approaches
FROM asteroids a
JOIN close_approach c
ON a.id = c.neo_reference_id
GROUP BY a.id, a.name
ORDER BY total_approaches DESC;
"""

run_query(query)
```

2. Average velocity of each asteroid over multiple approaches

```python
query = """
SELECT
    a.name,
    ROUND(AVG(c.relative_velocity_kmph),2) AS average_velocity
FROM asteroids a
JOIN close_approach c
ON a.id = c.neo_reference_id
GROUP BY a.id,a.name
ORDER BY average_velocity DESC;
"""

run_query(query)
```

3. List Top 10 Fastest Asteroids

```python
query = """
SELECT
    a.name,
    MAX(c.relative_velocity_kmph) AS max_velocity
FROM asteroids a
JOIN close_approach c
ON a.id=c.neo_reference_id
GROUP BY a.id,a.name
ORDER BY max_velocity DESC
LIMIT 10;
"""

run_query(query)
```

4. Hazardous asteroids that approached Earth more than 3 times

```python
query = """
SELECT
    a.name,
    COUNT(*) AS approaches
FROM asteroids a
JOIN close_approach c
ON a.id=c.neo_reference_id
WHERE a.is_potentially_hazardous_asteroid=True
GROUP BY a.id,a.name
HAVING COUNT(*)>3;
"""

run_query(query)
```

#5. Find the month with the most asteroid approaches

```python
query = """
SELECT
    MONTHNAME(close_approach_date) AS month,
    COUNT(*) AS total_approaches
FROM close_approach
GROUP BY MONTH(close_approach_date),
MONTHNAME(close_approach_date)
ORDER BY total_approaches DESC
LIMIT 1;
"""

run_query(query)
```

6. Asteroid with the fastest ever approach speed

```python
query = """
SELECT
    a.name,
    c.relative_velocity_kmph
FROM asteroids a
JOIN close_approach c
ON a.id=c.neo_reference_id
ORDER BY c.relative_velocity_kmph DESC
LIMIT 1;
"""

run_query(query)
```

7. Sort asteroids by maximum estimated diameter

```python
query = """
SELECT
    name,
    estimated_diameter_max_km
FROM asteroids
ORDER BY estimated_diameter_max_km DESC;
"""

run_query(query)
```

8. Asteroid whose closest approach is getting nearer over time

```python
query = """
SELECT
    a.name,
    c.close_approach_date,
    c.miss_distance_km
FROM asteroids a
JOIN close_approach c
ON a.id=c.neo_reference_id
ORDER BY a.name,c.close_approach_date;
"""

run_query(query)
```

9. Name, date and miss distance of closest approach

```python
query = """
SELECT
    a.name,
    c.close_approach_date,
    c.miss_distance_km
FROM asteroids a
JOIN close_approach c
ON a.id=c.neo_reference_id
WHERE c.miss_distance_km=
(
SELECT MIN(c2.miss_distance_km)
FROM close_approach c2
WHERE c2.neo_reference_id=a.id
)
ORDER BY c.miss_distance_km;
"""

run_query(query)
```

10. Velocity greater than 50,000 km/h

```python
query = """
SELECT
    MONTHNAME(close_approach_date) AS Month,
    COUNT(*) AS Total_Approaches
FROM close_approach
GROUP BY MONTH(close_approach_date),
MONTHNAME(close_approach_date)
ORDER BY MONTH(close_approach_date);
"""

run_query(query)
```

12. Asteroid with highest brightness

```python
query = """
SELECT
    name,
    absolute_magnitude_h
FROM asteroids
ORDER BY absolute_magnitude_h ASC
LIMIT 1;
"""

run_query(query)
```

13. Hazardous vs Non-Hazardous Asteroids

```python
query = """
SELECT
CASE
WHEN is_potentially_hazardous_asteroid=True
THEN 'Hazardous'
ELSE 'Non-Hazardous'
END AS Category,
COUNT(*) AS Total
FROM asteroids
GROUP BY is_potentially_hazardous_asteroid;
"""

run_query(query)
```

14. Passed closer than the Moon (<1 LD)

```python
query = """
SELECT
    a.name,
    c.close_approach_date,
    c.miss_distance_lunar
FROM asteroids a
JOIN close_approach c
ON a.id=c.neo_reference_id
WHERE c.miss_distance_lunar<1
ORDER BY c.miss_distance_lunar;
"""

run_query(query)
```

15. Came within 0.05 AU

```python
query = """
SELECT
    a.name,
    c.close_approach_date,
    c.astronomical
FROM asteroids a
JOIN close_approach c
ON a.id=c.neo_reference_id
WHERE c.astronomical<0.05
ORDER BY c.astronomical;
"""

run_query(query)
```
