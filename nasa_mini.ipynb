```python
asteroid_data = []
target_date = 200
API_Key= "EFeiAVwmT8TLI0Qw2nv3eN4etZX83OQC2ma1fkcG"
url = f"https://api.nasa.gov/neo/rest/v1/feed?start_date=2024-01-01&end_date=2024-01-07&api_key={API_Key}"

while(len(asteroid_data) < target_date):
    response = requests.get(url)
    data = response.json()
    near_earth_objects = data['near_earth_objects']
    
    for date, asteroids in near_earth_objects.items():
        for asteroid in asteroids:
            name = asteroid['name']
            diameter = asteroid['estimated_diameter']['meters']['estimated_diameter_max']
            is_hazardous = asteroid['is_potentially_hazardous_asteroid']
            asteroid_data.append({
                'date': date,
                'name': name,
                'diameter': diameter,
                'is_hazardous': is_hazardous
            })
    
    # Update the URL for the next page of results
    url = data['links']['next']
print(f"Collected {len(asteroid_data)} asteroids.")
```
